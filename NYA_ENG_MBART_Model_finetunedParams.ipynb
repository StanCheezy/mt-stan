```python
def normalise(text):
    text = re.sub('[{}]'.format(re.escape(",'$!@#$%^&*().\"")), ' ', text.strip())  # Remove special characters
    text = re.sub(r'\s+', ' ', text)        # Convert multiple whitespace into a single space
    return text

fieldnames = ['Two young guys with shaggy hair look at their hands while hanging out in the yard .', "Anyamata awiri acicepele atsitsi lopindika ayangana m'manja mwawo pomwe ali pabwalo."]
df = pd.read_csv('/content/drive/MyDrive/nya-eng.tsv', delimiter='\t', usecols=fieldnames, dtype='object', na_filter=False)
df = df.applymap(normalise)
print(df)
```

```python
from tqdm import tqdm, trange
from sklearn.model_selection import train_test_split

eng_data = []
nya_data = []

with open('/content/drive/MyDrive/nya-eng.tsv') as fp:
#     for line in tqdm(fp, total=3895992):
    for line in tqdm(fp, total=27291):
        line_data = line.rstrip().split('\t')
        a = line_data
        eng_data.append(a)
        nya_data.append(a)
        
        
en_data = []
ny_data = []
for sublist in eng_data:
    for item in sublist:
        ny_data.append(item + '\n')

        
for sublist in nya_data:
    for item in sublist:
        en_data.append(item + '\n')
        

total_test = 20000
en_train, en_subtotal, ny_train, ny_subtotal = train_test_split(
        en_data, ny_data, test_size=total_test, random_state=42)

en_test, en_val, ny_test, ny_val = train_test_split(
        en_subtotal, ny_subtotal, test_size=0.5, random_state=42)        



file_mapping = {
    'train.en_XX': en_train,
    'train.ny_XX': ny_train,
    'valid.en_XX': en_val,
    'valid.ny_XX': ny_val,
    'test.en_XX': en_test,
    'test.ny_XX': ny_test,

}
for k, v in file_mapping.items():
    with open(f'/content/preprocessed/{k}', 'w') as fp:
        fp.writelines(v)
```
